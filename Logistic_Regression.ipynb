Logistic Regression
It's a classification rather than regression algorithm.

It uses an 'S'-shaped curve instead of a straight line makes it a natural fit for dividing data into groups.

In this model, the probabilities describing the possible outcomes of a single trial are modeled using a logistic function

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import expit

# Create an array of x values from -6 to 6 with an increment of 0.02
x_values = np.arange(-6, 6.02, 0.02)

# Calculate sigmoid values using the expit function
sigmoid_values = expit(x_values)

# Plot the sigmoid curve
plt.plot(x_values, sigmoid_values, label='Sigmoid Function')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.legend()
plt.grid(True)
plt.show()
```
